```python
from random import randint,gauss,random
from math import cos,sqrt,exp

taw=0
tawPrime=0
epsilon=0.001

class Chromosome:

    def fitnessFunction(self):
        resultOfFunction = 0
        for index in range(0,len(self.objs)):
            resultOfFunction = resultOfFunction + (self.objs[index]**2)/4000
        temp = 1
        for index in range(0, len(self.objs)):
            temp = temp*cos(self.objs[index]/sqrt(index+1))
        return abs(resultOfFunction-temp+1)

    def __init__(self,objecs,type):
        if objecs != None:
            self.objs = objecs
        elif type == 1:
            tempObjs = list()
            for i in range(0,10):
                tempObjs.append(randint(-600,600))
            tempObjs.append(epsilon)
            self.objs = tempObjs
        elif type == 2:
            tempObjs = list()
            for i in range(0, 10):
                tempObjs.append(randint(-600, 600))
            for i in range(0, 10):
                tempObjs.append(epsilon)
            self.objs = tempObjs

        self.fitness = self.fitnessFunction()

    def mutation(self,type):
        if type==1:
            taw=1/sqrt(len(self.objs))
            self.objs[len(self.objs)-1]=self.objs[len(self.objs)-1]*exp(taw*gauss(0,1))
            if self.objs[len(self.objs)-1] < epsilon :
                self.objs[len(self.objs)-1]=epsilon
            for i in range (0,len(self.objs)):
                self.objs[i]=self.objs[i]+self.objs[10]*gauss(0,1)
        elif type==2:
            taw=1/sqrt(2*len(self.objs)/2)
            tawPrime=1/sqrt(2*sqrt(len(self.objs)/2))
            normalForOverall = gauss(0,1)
            for i in range(10,len(self.objs)):
                self.objs[i] = self.objs[i]*exp(tawPrime*normalForOverall+taw*gauss(0,1))
                if self.objs[i]<epsilon :
                    self.objs[i]=epsilon
            for i in range(0,len(self.objs)/2):
                self.objs[i] = self.objs[i]+self.objs[i+10]*gauss(0,1)
        self.fitness = self.fitnessFunction()

    def localIntermed(self,secPar):
        offspringData = list()
        for i in range(0,len(self.objs)):
            offspringData.append((self.objs[i]+secPar.objs[i])/2)
        offspring = Chromosome(offspringData,1)
        return offspring

    def globalIntermed(self,parents):
        index = 0
        offspringData = list()
        for i in range(len(self.objs)):
            temp = parents[index]
            temp = temp.objs[i]
            temp = temp + parents[index+1].objs[i]
            offspringData.append(temp/2)
            index = index+2
        offspring = Chromosome(offspringData)
        return offspring

    def localDiscrete(self,secPar):
        offspringData = list()
        for i in range(len(self.objs)):
            if random() > 0.5:
                offspringData.append(secPar.objs[i])
            else:
                offspringData.append(self.objs[i])
        offspring = Chromosome(offspringData,1)
        return offspring

    def globalDiscrete(self,parents):
        index = 0
        offspringData = list()
        for i in range(len(self.objs)):
            if random() >0.5:
                temp = parents[index].objs[i]
                offspringData.append(temp)
            else:
                temp = parents[index+1].objs[i]
                offspringData.append(temp)
            index = index+2
        offspring = Chromosome(offspringData)
        return offspring

class Population:
    def __init__(self,type):
        self.mem = list();
        for i in range(100):
            self.mem.append(Chromosome(None,type))

    def sort(list):
        for i in range(len(list)):
            min_idx = i
            for j in range(i + 1, len(list)):
                if list[min_idx].fitness > list[j].fitness:
                    min_idx = j
            list[i], list[min_idx] = list[min_idx], list[i]

    sort = staticmethod(sort)

    def parentSelector(self):
        return self.mem[randint(0,99)]

    def survivorSelector(self,childList,type):
        if type==1: #mo,lambda
            self.sort(childList)
            self.mem = childList[0:100]
        elif type==2: #mo+lambda
            childList.extend(self.mem)
            self.sort(childList)
            self.mem = childList[0:100]

def f1(): #only has one sigma , local intermed. , surv. mo,lamda
    p = Population(1)
    Population.sort(p.mem)
    while(p.mem[0].fitness > 1):
        children = list()
        for i in range(700):
            parent = list()
            parent.append(p.parentSelector())
            parent.append(p.parentSelector())
            children.append(parent[0].localIntermed(parent[1]))
            children[len(children)-1].mutation(1)
        p.survivorSelector(children,1)
        print(p.mem[0].objs)
        print(p.mem[0].fitness)
    print('only has one sigma , local intermed. , surv. mo,lamda')
    print('result: ', p.mem[0].fitness , p.mem[0].objs)
    print('________________________________________________')

def f2(): #only has one sigma , local intermed. , surv. mo+lamda
    p = Population(1)
    Population.sort(p.mem)
    while(p.mem[0].fitness > 1):
        children = list()
        for i in range(700):
            parent = list()
            parent.append(p.parentSelector())
            parent.append(p.parentSelector())
            children.append(parent[0].localIntermed(parent[1]))
            children[len(children)-1].mutation(1)
        p.survivorSelector(children,2)
        print(p.mem[0].objs)
        print(p.mem[0].fitness)
    print('only has one sigma , local intermed. , surv. mo+lamda')
    print('result: ', p.mem[0].fitness, p.mem[0].objs)
    print('________________________________________________')


def f3():  # only has one sigma , local disc. , surv. mo,lamda
    p = Population(1)
    Population.sort(p.mem)
    while (p.mem[0].fitness > 1.56):
        children = list()
        for i in range(700):
            parent = list()
            parent.append(p.parentSelector())
            parent.append(p.parentSelector())
            children.append(parent[0].localDiscrete(parent[1]))
            children[len(children) - 1].mutation(1)
        p.survivorSelector(children, 1)
        print(p.mem[0].objs)
        print(p.mem[0].fitness)
    print('only has one sigma , local disc. , surv. mo,lamda')
    print('result: ', p.mem[0].fitness, p.mem[0].objs)
    print('________________________________________________')


def f4():  # only has one sigma , local disc. , surv. mo+lamda
    p = Population(1)
    Population.sort(p.mem)
    while (p.mem[0].fitness > 1.33):
        children = list()
        for i in range(700):
            parent = list()
            parent.append(p.parentSelector())
            parent.append(p.parentSelector())
            children.append(parent[0].localDiscrete(parent[1]))
            children[len(children) - 1].mutation(1)
        p.survivorSelector(children, 2)
        print(p.mem[0].objs)
        print(p.mem[0].fitness)
    print('only has one sigma , local disc. , surv. mo+lamda')
    print('result: ', p.mem[0].fitness, p.mem[0].objs)
    print('________________________________________________')

def f5():  # only has one sigma , global intermed. , surv. mo,lamda
    p = Population(1)
    Population.sort(p.mem)
    while (p.mem[0].fitness > 1):
        children = list()
        for i in range(700):
            parent = list()
            parent.append(p.parentSelector())
            parent.append(p.parentSelector())
            children.append(parent[0].globalIntermed(parent[1]))
            children[len(children) - 1].mutation(1)
        p.survivorSelector(children, 1)
        print(p.mem[0].objs)
        print(p.mem[0].fitness)
    print('only has one sigma , global intermed. , surv. mo,lamda')
    print('result: ', p.mem[0].fitness, p.mem[0].objs)
    print('________________________________________________')

def f6():  # only has one sigma , global intermed. , surv. mo+lamda
    p = Population(1)
    Population.sort(p.mem)
    while (p.mem[0].fitness > 1):
        children = list()
        for i in range(700):
            parent = list()
            parent.append(p.parentSelector())
            parent.append(p.parentSelector())
            children.append(parent[0].globalIntermed(parent[1]))
            children[len(children) - 1].mutation(1)
        p.survivorSelector(children, 2)
        print(p.mem[0].objs)
        print(p.mem[0].fitness)
    print('only has one sigma , global intermed. , surv. mo+lamda')
    print('result: ', p.mem[0].fitness, p.mem[0].objs)
    print('________________________________________________')

def f7():  # only has one sigma , global disc. , surv. mo,lamda
    p = Population(1)
    Population.sort(p.mem)
    while (p.mem[0].fitness > 1.45):
        children = list()
        for i in range(700):
            parent = list()
            parent.append(p.parentSelector())
            parent.append(p.parentSelector())
            children.append(parent[0].globalDiscrete(parent[1]))
            children[len(children) - 1].mutation(1)
        p.survivorSelector(children, 1)
        print(p.mem[0].objs)
        print(p.mem[0].fitness)
    print('only has one sigma , global disc. , surv. mo,lamda')
    print('result: ', p.mem[0].fitness, p.mem[0].objs)
    print('________________________________________________')

def f8():  # only has one sigma , global disc. , surv. mo+lamda
    p = Population(1)
    Population.sort(p.mem)
    while (p.mem[0].fitness > 1):
        children = list()
        for i in range(700):
            parent = list()
            parent.append(p.parentSelector())
            parent.append(p.parentSelector())
            children.append(parent[0].globalDiscrete(parent[1]))
            children[len(children) - 1].mutation(1)
        p.survivorSelector(children, 2)
        print(p.mem[0].objs)
        print(p.mem[0].fitness)
    print('only has one sigma , global disc. , surv. mo+lamda')
    print('result: ', p.mem[0].fitness, p.mem[0].objs)
    print('________________________________________________')


def f9():  # multi sigma , local intermed. , surv. mo,lamda
    p = Population(1)
    Population.sort(p.mem)
    while (p.mem[0].fitness > 1):
        children = list()
        for i in range(700):
            parent = list()
            parent.append(p.parentSelector())
            parent.append(p.parentSelector())
            children.append(parent[0].localIntermed(parent[1]))
            children[len(children) - 1].mutation(1)
        p.survivorSelector(children, 1)
        print(p.mem[0].objs)
        print(p.mem[0].fitness)
    print('multi sigma , local intermed. , surv. mo,lamda')
    print('result: ', p.mem[0].fitness, p.mem[0].objs)
    print('________________________________________________')


def f10():  # multi sigma , local intermed. , surv. mo+lamda
    p = Population(2)
    Population.sort(p.mem)
    while (p.mem[0].fitness > 1):
        children = list()
        for i in range(700):
            parent = list()
            parent.append(p.parentSelector())
            parent.append(p.parentSelector())
            children.append(parent[0].localIntermed(parent[1]))
            children[len(children) - 1].mutation(1)
        p.survivorSelector(children, 2)
        print(p.mem[0].objs)
        print(p.mem[0].fitness)
    print('multi sigma , local intermed. , surv. mo+lamda')
    print('result: ', p.mem[0].fitness, p.mem[0].objs)
    print('________________________________________________')


def f11():  # multi sigma , local disc. , surv. mo,lamda
    p = Population(2)
    Population.sort(p.mem)
    while (p.mem[0].fitness > 1):
        children = list()
        for i in range(700):
            parent = list()
            parent.append(p.parentSelector())
            parent.append(p.parentSelector())
            children.append(parent[0].localDiscrete(parent[1]))
            children[len(children) - 1].mutation(1)
        p.survivorSelector(children, 1)
        print(p.mem[0].objs)
        print(p.mem[0].fitness)
    print('multi sigma , local disc. , surv. mo,lamda')
    print('result: ', p.mem[0].fitness, p.mem[0].objs)
    print('________________________________________________')


def f12():  # multi sigma , local disc. , surv. mo+lamda
    p = Population(2)
    Population.sort(p.mem)
    while (p.mem[0].fitness > 1):
        children = list()
        for i in range(700):
            parent = list()
            parent.append(p.parentSelector())
            parent.append(p.parentSelector())
            children.append(parent[0].localDiscrete(parent[1]))
            children[len(children) - 1].mutation(1)
        p.survivorSelector(children, 2)
        print(p.mem[0].objs)
        print(p.mem[0].fitness)
    print('multi sigma , local disc. , surv. mo+lamda')
    print('result: ', p.mem[0].fitness, p.mem[0].objs)
    print('________________________________________________')


def f13():  # multi sigma , global intermed. , surv. mo,lamda
    p = Population(2)
    Population.sort(p.mem)
    while (p.mem[0].fitness > 1):
        children = list()
        for i in range(700):
            parent = list()
            parent.append(p.parentSelector())
            parent.append(p.parentSelector())
            children.append(parent[0].globalIntermed(parent[1]))
            children[len(children) - 1].mutation(1)
        p.survivorSelector(children, 1)
        print(p.mem[0].objs)
        print(p.mem[0].fitness)
    print('multi sigma , global intermed. , surv. mo,lamda')
    print('result: ', p.mem[0].fitness, p.mem[0].objs)
    print('________________________________________________')


def f14():  # multi sigma , global intermed. , surv. mo+lamda
    p = Population(2)
    Population.sort(p.mem)
    while (p.mem[0].fitness > 1):
        children = list()
        for i in range(700):
            parent = list()
            parent.append(p.parentSelector())
            parent.append(p.parentSelector())
            children.append(parent[0].globalIntermed(parent[1]))
            children[len(children) - 1].mutation(1)
        p.survivorSelector(children, 2)
        print(p.mem[0].objs)
        print(p.mem[0].fitness)
    print('multi sigma , global intermed. , surv. mo+lamda')
    print('result: ', p.mem[0].fitness, p.mem[0].objs)
    print('________________________________________________')


def f15():  # multi sigma , global disc. , surv. mo,lamda
    p = Population(2)
    Population.sort(p.mem)
    while (p.mem[0].fitness > 1):
        children = list()
        for i in range(700):
            parent = list()
            parent.append(p.parentSelector())
            parent.append(p.parentSelector())
            children.append(parent[0].globalDiscrete(parent[1]))
            children[len(children) - 1].mutation(1)
        p.survivorSelector(children, 1)
        print(p.mem[0].objs)
        print(p.mem[0].fitness)
    print('multi sigma , global disc. , surv. mo,lamda')
    print('result: ', p.mem[0].fitness, p.mem[0].objs)
    print('________________________________________________')


def f16():  # multi sigma , global disc. , surv. mo+lamda
    p = Population(2)
    Population.sort(p.mem)
    while (p.mem[0].fitness > 1):
        children = list()
        for i in range(700):
            parent = list()
            parent.append(p.parentSelector())
            parent.append(p.parentSelector())
            children.append(parent[0].globalDiscrete(parent[1]))
            children[len(children) - 1].mutation(1)
        p.survivorSelector(children, 2)
        print(p.mem[0].objs)
        print(p.mem[0].fitness)
    print('multi sigma , global disc. , surv. mo+lamda')
    print('result: ', p.mem[0].fitness, p.mem[0].objs)
    print('________________________________________________')

for i in range(10):
    f1()

for i in range(10):
    f2()

for i in range(10):
    f3()

for i in range(10):
    f4()

for i in range(10):
    f5()

for i in range(10):
    f6()

for i in range(10):
    f7()

for i in range(10):
    f8()

for i in range(10):
    f9()

for i in range(10):
    f10()

for i in range(10):
    f11()

for i in range(10):
    f12()

for i in range(10):
    f13()

for i in range(10):
    f14()

for i in range(10):
    f15()

for i in range(10):
    f16()
```
